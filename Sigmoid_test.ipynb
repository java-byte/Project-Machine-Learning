```python
X_train,X_test,y_train,y_test = train_test_split(x_train,y_train,test_size=0.33,random_state=42)

X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

## Linear Regression

```python
## Linear Regression
from sklearn.linear_model import LinearRegression

X_train_linear = X_train
y_train_linear = y_train
X_test_linear = X_test
y_test_linear = y_test
regr_linear = LinearRegression().fit(X_train_linear, y_train_linear)
regr_linear.score(X_train_linear, y_train_linear)
```

```python
## Root Mean square error

prediction_train_linear = np.array(regr_linear.predict(X_train_linear),dtype=np.int64).reshape(1,-1)
prediction_test_linear = np.array(regr_linear.predict(X_test_linear),dtype=np.int64).reshape(1,-1)

y_train_linear = np.array(y_train_linear).reshape(1,-1)
y_test_linear = np.array(y_test_linear).reshape(1,-1)

print("Training accuracy: (RMSE) ",np.sqrt(mean_squared_error(prediction_train_linear,y_train_linear)))
print("Test accuracy: (RMSE) ",np.sqrt(mean_squared_error(prediction_test_linear,y_test_linear)))
```

```python
# Random Forest

X_train_rf = X_train
y_train_rf = y_train
X_test_rf = X_test
y_test_rf = y_test

regr = RandomForestRegressor(max_depth=2, random_state=0,n_estimators=100)
regr.fit(X_train_rf,y_train_rf)
```

```python
X_train_rf.shape,y_train_rf.shape,X_test_rf.shape,y_test_rf.shape
```

```python
prediction_train_rf = np.array(regr.predict(X_train_rf),dtype=np.int64).reshape(-1,1)
prediction_test_rf = np.array(regr.predict(X_test_rf),dtype=np.int64).reshape(-1,1)

y_train_rf = np.array(y_train_rf).reshape(-1,1)
y_test_rf = np.array(y_test_rf).reshape(-1,1)

print(prediction_train_rf.shape,y_train_rf.shape,prediction_test_rf.shape,y_test_rf.shape)

print("Training accuracy: (RMSE) ",np.sqrt(mean_squared_error(prediction_train_rf,y_train_rf)))
print("Test accuracy: (RMSE) ",np.sqrt(mean_squared_error(prediction_test_rf,y_test_rf)))
```

```python
## Support Vector Machine

from sklearn.svm import SVR

X_train_sv = X_train
y_train_sv = y_train
X_test_sv = X_test
y_test_sv = y_test

clf_sv = SVR(C=1, epsilon=0.2)
clf_sv.fit(X_train_sv,y_train_sv)
```
