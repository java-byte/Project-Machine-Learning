```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sb
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
```

```python
train_df = pd.read_csv('training_data.csv')

train_df.head()
```

```python
train_df.describe()
```

```python
x_train = train_df.iloc[:,1:]
x_train.head()
```

```python
y_train = train_df.iloc[:,-1]
y_train
```

```python
x_train.shape,y_train.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(x_train,y_train,test_size=0.33,random_state=42)

X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
## Mean Absolute Percentage Error (MAPE)
from sklearn.utils import check_array

def mean_absolute_percentage_error(y_pred,y_true):
    #y_pred = np.transpose(y_pred)
    #y_true = np.transpose(y_true)
    #y_true, y_pred = check_array(y_true,y_pred)
    
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
y_train = np.array(y_train).reshape(-1,1)
y_test = np.array(y_test).reshape(-1,1)
y_train.shape,y_test.shape
```

```python
# Random Forest

regr = RandomForestRegressor(max_depth=2, random_state=0,n_estimators=100)
regr.fit(X_train,y_train)
prediction_train = np.array(regr.predict(X_train)).reshape(-1,1)
prediction_test = np.array(regr.predict(X_test)).reshape(-1,1)
print("Training accuracy: ",mean_absolute_percentage_error(prediction_train,y_train))
print("Test accuracy: ",mean_absolute_percentage_error(prediction_test,y_test))
```

```python
## Mean square error
prediction_train = np.array(regr.predict(X_train)).reshape(1,-1)
prediction_test = np.array(regr.predict(X_test)).reshape(1,-1)

y_train = np.array(y_train).reshape(1,-1)
y_test = np.array(y_test).reshape(1,-1)

print("Training accuracy: ",mean_squared_error(prediction_train,y_train))
print("Test accuracy: ",mean_squared_error(prediction_test,y_test))
```
